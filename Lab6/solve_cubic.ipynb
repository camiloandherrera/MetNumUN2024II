```python
import cmath

def solve_cubic(b, c, d):

    a = 1

    Delta_0 = b**2 -3*a*c
    Delta_1 = 2*b**3 - 9*a*b*c + 27*a**2*d

    Delta_3 = Delta_1**2 - 4*Delta_0**3
    Delta_4 = Delta_1 + cmath.sqrt(Delta_3)

    if(Delta_4 == 0):
      Delta_4 = Delta_1 - cmath.sqrt(Delta_3)

    C = complex(Delta_4 /2)**(1/3)

    xi = - complex(1)**(1/3) * (1j**(2/3))

    if(C == 0):
      roots = [-b/(3*a), -b/(3*a), -b/(3*a)]
    else:
      roots = []
      for k in range(3):
        root = (-1/(3*a)) * (b + xi**k * C + Delta_0 / (xi**k * C))
        roots.append(root)

    return roots
```
